```python
# 2) Get Hugging Face token securely
import os
from google.colab import userdata

os.environ["HF_TOKEN"] = userdata.get('HF_TOKEN')
```

```python
# 3) Use AISuite with Hugging Face provider
import aisuite as ai

client = ai.Client()  # will pick up HF_TOKEN from env

messages = [
    {"role": "system", "content": "Be concise."},
    {"role": "user", "content": "Tell me a Computing joke."},
]

# Use a model repo id from the Hub (change as you like)
resp = client.chat.completions.create(
    model="huggingface:meta-llama/Llama-3.2-1B-Instruct",
    messages=messages,
    temperature=0.7,
)

print(resp.choices[0].message.content)
```
